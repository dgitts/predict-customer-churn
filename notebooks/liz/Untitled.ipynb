```python
#load the required data from the excel file onto pandas dataframe.
df=pd.read_excel("C:\\Users\\regin\\Desktop\\lale\\datascienceexel\\DSII.xlsx",header=None)
df.head()
```

```python
new_header = df.iloc[0] #grab the first row for the header
df = df[1:] #take the data under the header row
df.columns = new_header
df.head()
```

```python
def missing_values_table(df):
        mis_val = df.isnull().sum()
        mis_val_percent = 100 * df.isnull().sum() / len(df)
        mis_val_table = pd.concat([mis_val, mis_val_percent], axis=1)
        mis_val_table_ren_columns = mis_val_table.rename(
        columns = {0 : 'Missing Values', 1 : '% of Total Values'})
        mis_val_table_ren_columns = mis_val_table_ren_columns[
            mis_val_table_ren_columns.iloc[:,1] != 0].sort_values(
        '% of Total Values', ascending=False).round(1)
        print ("Your selected dataframe has " + str(df.shape[1]) + " columns.\n"      
            "There are " + str(mis_val_table_ren_columns.shape[0]) +
              " columns that have missing values.")
        return mis_val_table_ren_columns

    
missing_values_table(df)
```

```python
#drop null/NaN 
df=df.dropna()
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
df.columns
```

```python
df=df.drop("Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_2",axis='columns')
```

```python
df= df.drop("Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_1",axis='columns')
df.columns
```

```python
le_Attrition_Flag=LabelEncoder()
le_Gender=LabelEncoder()
le_Education_Level=LabelEncoder()
le_Marital_Status=LabelEncoder()
le_Income_Category=LabelEncoder()
le_Card_Category=LabelEncoder()
```

```python
df['Attrition_Flag_n']=le_Attrition_Flag.fit_transform(df['Attrition_Flag'])
df['Gender_n']=le_Gender.fit_transform(df['Gender'])
df['Education_Level_n']=le_Education_Level.fit_transform(df['Education_Level'])
df['Marital_Status_n']=le_Marital_Status.fit_transform(df['Marital_Status'])
df['Income_Category_n']=le_Income_Category.fit_transform(df['Income_Category'])
df['Card_Category_n']=le_Card_Category.fit_transform(df['Card_Category'])
df.head()
```

```python
df_n=df.drop(['Attrition_Flag','Gender','Education_Level','Marital_Status','Income_Category','Card_Category'], axis='columns')
df_n.head()
```

```python
attrition_df=df['Attrition_Flag_n']
df=df.drop('Attrition_Flag_n',axis='columns')
attrition_df.head()
```

```python
from sklearn.model_selection import train_test_split
# Split dataset into training set and test set
df_ntrain, df_ntest, attrition_dftrain, attrition_dftest = train_test_split(df_n, attrition_df, test_size=0.2, random_state=1)
```

```python
from sklearn import tree

# Create Decision Tree classifer object
clf = tree.DecisionTreeClassifier()

# Train Decision Tree Classifer
clf = clf.fit(df_ntrain,attrition_dftrain)

#Predict the response for test dataset
y_pred = clf.predict(df_ntest)
```

```python
# Model Accuracy, how often is the classifier correct?
from sklearn import metrics #Import scikit-learn metrics module for accuracy calculation
print("Accuracy:",metrics.accuracy_score(attrition_dftest, y_pred))
```
